```python
#Replacing the string columns with int values:
#No replaced with 0
#Yes replaced with 1
# Any other string value in the column with 2

data2['Partner'].replace(['No','Yes'],[0,1],inplace=True)
data2['Dependents'].replace(['No','Yes'],[0,1],inplace=True)
data2['PhoneService'].replace(['No','Yes'],[0,1],inplace=True)
data2['Churn'].replace(['No','Yes'],[0,1],inplace=True)
data2['PaperlessBilling'].replace(['No','Yes'],[0,1],inplace=True)
data2['MultipleLines'].replace(['No','Yes', 'No phone service'],[0,1, 2],inplace=True)
data2['OnlineSecurity'].replace(['No', 'Yes', 'No internet service'],[0,1,2],inplace=True)
data2['TechSupport'].replace(['No', 'Yes', 'No internet service'],[0,1,2],inplace=True)
data2['StreamingTV'].replace(['No', 'Yes', 'No internet service'],[0,1,2],inplace=True)
data2['StreamingMovies'].replace(['No', 'Yes', 'No internet service'],[0,1,2],inplace=True)
data2['OnlineBackup'].replace(['No', 'Yes', 'No internet service'],[0,1,2],inplace=True)
data2['DeviceProtection'].replace(['No', 'Yes', 'No internet service'],[0,1,2],inplace=True)
data2['Contract'].replace(['Month-to-month', 'One year', 'Two year'],[0,1,2],inplace=True)
data2['gender'].replace(['Male', 'Female'],[0,1],inplace=True)
```

```python
#Run Logistic regression model by including the features which had high positive and negative coorelation with churn rate
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

feature_cols = [ 'SeniorCitizen', 'Dependents' , 'tenure',  
               'Contract', 'OnlineSecurity', 'PaperlessBilling', 'gender', 'TechSupport', 
             'OnlineBackup', 'DeviceProtection']
                
X = data2[feature_cols]
y = data2.Churn

logreg.fit(X,y)
pred = logreg.predict(X)
```

Features that are important predictors of churn rate are SeniorCitizen, Having dependents, gender, tenure, contract 
duration, Online security, Paperless billing, Tech support, Device Protection and Online backup. 

```python
#Logistic Regression
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train.shape, X_test.shape


logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print("{0} / {1} correct".format(np.sum(y_test == y_pred), len(y_test)))
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred) 
```

Logistic Regression model has accuracy of 79%

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
conf = confusion_matrix(y_test,y_pred)
print(conf)
```
